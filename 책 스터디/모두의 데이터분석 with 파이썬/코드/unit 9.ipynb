```python
import csv
from matplotlib import pyplot as plt

f=open("gender.csv")
data=csv.reader(f)

next(data)
m=[]
fe=[]

name=input("지역을 입력해주세요 : ")

for i in data:
    if name in i[0]:
        for j in range(0,101):
            try:
                
                m.append(-int(i[j+3]))
                fe.append(int(i[-(j+1)]))
            except:
                m.append(-int(i[j+3].replace(",","")))
                fe.append(int(i[-(j+1)].replace(",","")))
            print(len(m),len(fe))
        break
        
fe.reverse()
print((m),(fe))
print(len(m),len(fe))

plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title("%s 의 남녀 비율"%(name))

plt.style.use('ggplot')

plt.barh(range(101),m,color="blue",label='남')
plt.barh(range(101),fe,color="red",label='여')
plt.xlabel("명")
plt.ylabel("나이")
plt.legend()
plt.show()

f.close       
```

```python
import csv
 
f = open('gender.csv')
data = csv.reader(f)
size = []
name = input('찾고 싶은 지역의 이름을 알려주세요 : ')
for row in data :
    if name in row[0] :
        m = 0
        f = 0
        for i in range(101) :
            try:
                m += int(row[i+3])    # 3 ~ 103번까지 데이터 더하기
                f += int(row[i+106])  # 106 ~ 206번까지 데이터 더하기
            except:
                m += int(row[i+3].replace(",",""))
                f += int(row[i+106].replace(",",""))
                
                
        break
        
size.append(m)                    # 남성 합계 데이터 리스트에 추가
size.append(f)                    # 여성 합계 데이터 리스트에 추가
print(size)

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels=['남', '여'], autopct='%.1f%%', colors=color, startangle=90)
plt.title(name+' 지역의 남녀 성별 비율')
plt.show()
```
