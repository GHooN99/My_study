```python
from matplotlib import pyplot as plt
import random as ran            #랜덤 모듈 

dice=[]
num=int(input())

for i in range(num):
    dice.append(ran.randint(1,6)) 
    
plt.hist(dice,bins=6)                 #가로축 구간 개수 설정 
plt.show()
```

```python
from matplotlib import pyplot as plt
import csv 

f=open("seoul.csv")
data=csv.reader(f)
header=next(data)

temp8=[]
temp1=[]

for row in data:
    if row[-1]!="":       #데이터가 있으면 
        
        date=row[0].split("-")    #date 리스트에 날짜저장
        
        if date[1]=="08":            # 8월이면 추가
            temp8.append(float(row[-1]))   
        if date[1]=="01":
            temp1.append(float(row[-1]))
            
plt.hist(temp8,bins=100,color="red")
plt.hist(temp1,bins=100,color="skyblue")
plt.show()
```

```python
from matplotlib import pyplot as plt
import csv 

f=open("seoul.csv")
data=csv.reader(f)
header=next(data)

month=[[],[],[],[],[],[],[],[],[],[],[],[]]             # 월 을 나타내는 이중리스트  

for row in data:
    if row[-1]!="":       #데이터가 있으면 
        date=row[0].split("-")    #date 리스트에 날짜저장
        month[int(date[1])-1].append(float(row[-1]))       #이중리스트 인덱스 > 월-1 에 기온요소 추가
plt.boxplot(month)
plt.show()
                                     
```

```python
from matplotlib import pyplot as plt
import csv 

f=open("seoul.csv")
data=csv.reader(f)
header=next(data)

month=[[],[],[],[],[],[],[],[],[],[],[],[]]             # 월 을 나타내는 이중리스트  

for row in data:
    if row[-1]!="":       #데이터가 있으면 
        date=row[0].split("-")
        month[int(date[1])-1].append(float(row[-1]))
                                     
plt.boxplot(month)
plt.show()
                                     
```
