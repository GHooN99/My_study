```python
#split 함수 응용 과 시각화
from matplotlib import pyplot as plt
import csv 

f=open("seoul.csv")
data=csv.reader(f)
header=next(data)

temp=[]

for row in data:
    if row[-1]!="":       #데이터가 있으면 
        
        date=row[0].split("-")    #date 리스트에 날짜저장
        
        if date[1]=="08":            # 8월이면 추가
            temp.append(float(row[-1]))   
            
            
#데이터 시각화
            
plt.title("Aug temp")       
plt.plot(temp,label="temp")
plt.legend()

plt.show()
    
```
