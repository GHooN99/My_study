```python
from matplotlib import pyplot as plt
import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]
xaxis=[]
yaxis=[]

for i in data:
    date=i[0].split('-')
    if i[-1]!='':
        if date[1]=='03'and date[2]=="16":
            yaxis.append(float(i[-1]))
            xaxis.append(int(date[0]))

result=[xaxis,yaxis]
print(result)
print(len(result))

plt.plot(xaxis,yaxis,color="blue",label="temp",ls="dotted")
plt.xticks(xaxis)
plt.title("Hello")
plt.show()

f.close
```
