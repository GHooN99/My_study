```python
import csv
from matplotlib import pyplot as plt

f=open("age.csv")
data=csv.reader(f)

next(data)
x=[]

name=input("지역을 입력해주세요 : ")

for i in data:
    try:
        if name in i[0]:
            for j in i[3:]:
                x.append(int(j))
    except:
        name=input("지역을 입력해주세요 : ")
        
plt.rc('font',family='Malgun Gothic')
plt.title("%s 의 인구 구조"%(name))
plt.style.use('ggplot')
plt.plot(x,color="blue")
plt.xlabel("나이")
plt.ylabel("명")
plt.show()

f.close       
```

```python
import csv
from matplotlib import pyplot as plt

f=open("gender.csv")
data=csv.reader(f)

next(data)
m=[]
fe=[]

name=input("지역을 입력해주세요 : ")

for i in data:
    try:
        if name in i[0]:
            for j in range(0,101):
                
                m.append(-int(i[j+3]))
                fe.append(int(i[-(j+1)]))
    except:
        name=input("지역을 입력해주세요 : ")
fe.reverse()

plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title("%s 의 인구 구조"%(name))
plt.style.use('ggplot')
plt.barh(range(101),m,color="blue",label='남')
plt.barh(range(101),fe,color="red",label='여')
plt.xlabel("명")
plt.ylabel("나이")
plt.legend()
plt.show()

f.close       
```
