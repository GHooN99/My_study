```python
import csv
from matplotlib import pyplot as plt

f=open("age.csv")
data=csv.reader(f)

next(data)
x=[]
for i in data:
    try:
        if "과천시 부림동" in i[0]:
            for j in i[3:]:
                x.append(int(j))
    except:
        pass
plt.rc('font',family='Malgun Gothic')
plt.title("과천시 부림동 인구 구조")
plt.style.use('ggplot')
plt.plot(x,color="blue")
plt.xlabel("나이")
plt.ylabel("명")
plt.show()

f.close       
```

```python
import csv
from matplotlib import pyplot as plt

f=open("age.csv")
data=csv.reader(f)

next(data)
x=[]

name=input("지역을 입력해주세요 : ")

for i in data:
    try:
        if name in i[0]:
            for j in i[3:]:
                x.append(int(j))
                
plt.rc('font',family='Malgun Gothic')
plt.title("%s 의 인구 구조"%(name))
plt.style.use('ggplot')
plt.plot(x,color="blue")
plt.xlabel("나이")
plt.ylabel("명")
plt.show()

f.close       
```
