```python
import csv 
import matplotlib.pyplot as plt

f=open("subwaytime.csv")

data=csv.reader(f)
result=[]
for i in range(2):
    next(data)


for i in data:
    
    i[4:]=map(int,i[4:])
    result.append(sum(i[10:15:2]))

result.sort()
plt.bar(range(len(result)),result)
plt.style.use("ggplot")
plt.show()
    
f.close
```

```python
import csv 
import matplotlib.pyplot as plt

f=open("subwaytime.csv")

data=csv.reader(f)
result=[]
station=[]
for i in range(2):
    next(data)


for i in data:
    
    i[4:]=map(int,i[4:])
    result.append(sum(i[10:15:2]))
    station.append(i[1]+" "+i[3])

idx=result.index(max(result))
st=station[idx]

print("출근시간 가장 높은 승차인원")
print(st,str(max(result))+" 명")


f.close
```

```python
import csv 
import matplotlib.pyplot as plt

f=open("subwaytime.csv")

data=csv.reader(f)
result=[]
station=[]
for i in range(2):
    next(data)

t=int(input("몇시의 승차인원?(24시) : "))
for i in data:
    
    i[4:]=map(int,i[4:])
    result.append(i[(4-t)*2+4])
    station.append(i[1]+" "+i[3])

idx=result.index(max(result))
st=station[idx]

print("%d 시간에 가장 높은 승차인원"%(t))
print(st,str(max(result))+" 명")


f.close
```

```python
import csv 
import matplotlib.pyplot as plt

f=open("subwaytime.csv")

data=csv.reader(f)
maxresult=[0]*24
station=['']*24

for i in range(2):
    next(data)

for i in data:
    
    i[4:]=map(int,i[4:])
    
    for j in range(24):
        a=i[j*2+4]
        if a>maxresult[j]:
            maxresult[j]=a
            station[j]=i[1]+' '+i[3]+' '+str(j+4)+"시"


print(maxresult)
print(station)

plt.rc("font",family="Malgun Gothic")

plt.title("시간별 최대 승차 인원")
plt.bar(range(24),maxresult)
plt.xticks(range(24),station,rotation=90)
plt.ylabel("명")
plt.xlabel("역")
plt.show()


f.close
```

```python
import csv 
import matplotlib.pyplot as plt

f=open("subwaytime.csv")

data=csv.reader(f)
ins=[0]*24
outs=[0]*24


for i in range(2):
    next(data)

for i in data:
    
    i[4:]=map(int,i[4:])
    
    for j in range(24):
        ins[j]+=i[j*2+4]
        outs[j]+=i[j*2+5]
        

plt.rc("font",family="Malgun Gothic")
plt.plot(ins,label="승차")
plt.plot(outs,label="하차")
plt.xticks(range(24),range(4,28))
plt.title("시간별 승하차 인원")
plt.xlabel("시")
plt.ylabel("인원(천만명)")
plt.legend()
plt.show()

  
```
