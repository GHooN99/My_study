```python
from matplotlib import pyplot as plt
import random as ran
a=[]
for i in range(1000):
    a.append(ran.randint(1,6))

plt.hist(a,bins=6)
plt.show()
```

```python
import csv
f=open("seoul2.csv")
data=csv.reader(f)
next(data)
result8=[]
result1=[]
for i in data:
    mon=i[0].split("-")
    try:
        if mon[1]=="08":
            result8.append(float(i[-1]))
        if mon[1]=="01":
            result1.append(float(i[-1]))
    except:
        pass
plt.hist(result8,bins=100,color="red",label="Aug")
plt.hist(result1,bins=100,color="blue",label="Jan")
plt.legend()

plt.show()
```

```python
import csv
f=open("seoul2.csv")
data=csv.reader(f)
next(data)
result=[]
for i in range(12):
    result.append([])
    
for i in data:
    mon=i[0].split("-")
    try:
        result[int(mon[1])-1].append(float(i[-1]))
    except:
        pass

plt.boxplot(result)

plt.show()

            
        
```
