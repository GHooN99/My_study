```python
from matplotlib import pyplot as plt      # matplotlib 모듈 불러오기 plt 로 사용

plt.plot([1,2,3,4],[10,2,30,5],label="number1",color="red")             #plot 함수 (x축리스트,y축 리스트) ,라벨 설정,색상설정
plt.plot([2,3,4,5],[30,10,20,4],label="number2",color="blue")

plt.title("Hello")                          #제목 함수
plt.legend()                             #라벨 표시 함수

plt.show()    
```

```python
from matplotlib import pyplot as plt      # matplotlib 모듈 불러오기 plt 로 사용

plt.plot([1,2,3,4],[10,2,30,5],label="dashed",color="r",ls="--")             #LineStyle =ls
plt.plot([2,3,4,5],[30,10,20,4],label="dotted",color="g",ls=":")

plt.title("Hello")                          #제목 함수
plt.legend()                             #라벨 표시 함수

plt.show()    
```
