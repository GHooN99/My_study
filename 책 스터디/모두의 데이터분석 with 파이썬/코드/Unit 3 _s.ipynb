```python
#최고 기온 ,날짜 찾기

import csv

f=open("seoul.csv")  

data=csv.reader(f)        

header=next(data)             #헤더 저장하기 

temp=[]                         # 최고 기온 저장 리스트  
date=[]                         # 날짜 저장 리스트

for row in data:               #헤더 생략 후 출력 
    
    if row[-1]!='':               # 빈데이터 처리 
        
        temp.append(float(row[-1]))   # 최고 기온 리스트에 추가 (형변환 필요)
        date.append(row[0])           # 날짜 리스트 추가
    

max_temp=max(temp)                   # 최고 기온 찾기 

max_idx=temp.index(max_temp)         # 최고 기온 날짜 찾기 
max_date=date[max_idx]

print(str(max_temp)+"도",max_date)   #결과 출력


f.close()


```

```python
#최고 최저 기온, 일교차 ,날짜 찾기 (예외처리)

import csv

f=open("seoul.csv")  

data=csv.reader(f)        

header=next(data)             #헤더 저장하기 

maxtemp=[]                         # 최고 기온 저장 리스트  
mintemp=[]                         #최저 기온 
cross=[]                           #일교차 
date=[]                         # 날짜 저장 리스트

for row in data:               #헤더 생략 후 출력 
    
   # if row[-1]!='':               # 빈데이터 처리 대신에

    try:                          #우선 시도 
        
        mintemp.append(float(row[-2]))  #최저기온 리스트 추가
        maxtemp.append(float(row[-1]))   # 최고 기온 리스트에 추가 (형변환 필요)
        
        cross.append(float(row[-1])-float(row[-2]))   #일교차 리스트
        
        date.append(row[0])           # 날짜 리스트 추가
    
    except:                      #예외 (에러가 생기면)
        
        pass                  

#최고기온
max_temp=max(maxtemp)                   # 최고 기온 찾기 

max_idx=maxtemp.index(max_temp)         # 최고 기온 날짜 찾기 
max_date=date[max_idx]

print(str(max_temp)+"도",max_date)   #결과 출력

#최저기온
min_temp=min(mintemp)                   # 최저 기온 찾기 

min_idx=mintemp.index(min_temp)         # 최저 기온 날짜 찾기 
min_date=date[min_idx]

print(str(min_temp)+"도",min_date)   #결과 출력


f.close()


```

```python
#최고 최저 기온, 일교차 ,날짜 찾기 (예외처리)(함수)

import csv



def find(maxormin,templist,date):                #반복된작업 >> 함수 선언 
    
    if maxormin:            #최대값 찾기
        
        find_temp=max(templist)               # 최고 기온 찾기 
        
    else:                  #최소값 찾기
        
        find_temp=min(templist)
        
    find_idx=templist.index(find_temp)         # 최고/최저 기온 날짜 찾기 
    find_date=date[find_idx]
    
    return find_temp,find_date                 #최고/최저 기온 , 날짜 동시 반환 
    
    
    
f=open("seoul.csv")  

data=csv.reader(f)        

header=next(data)             #헤더 저장하기 

maxtemp=[]                         # 최고 기온 저장 리스트  
mintemp=[]                         #최저 기온 
cross=[]                           #일교차 
date=[]                         # 날짜 저장 리스트

for row in data:               #헤더 생략 후 출력 
    
   # if row[-1]!='':               # 빈데이터 처리 대신에

    try:                          #우선 시도 
        
        mintemp.append(float(row[-2]))  
        maxtemp.append(float(row[-1]))   # 최고 기온 리스트에 추가 (형변환 필요)
        
        cross.append(float(row[-1])-float(row[-2]))
        
        date.append(row[0])           # 날짜 리스트 추가
    
    except:                      #예외 (에러가 생기면)
        
        pass                  

    
    
maxtemp,maxdate =find(True,maxtemp,date)                   #동시 반환된 값 각각 할당 

mintemp,mindate=find(False,mintemp,date)

crosstemp,crossdate=find(True,cross,date)


print("서울의 역대 최고 기온은 "+str(maxdate)+"에 "+str(maxtemp)+"도 입니다.")
print("서울의 역대 최저 기온은 "+str(mindate)+"에 "+str(mintemp)+"도 입니다.")
print("서울의 역대 최고 일교차는 "+str(crossdate)+"에 "+str(crosstemp)+"도 입니다.")

    

f.close()


```
