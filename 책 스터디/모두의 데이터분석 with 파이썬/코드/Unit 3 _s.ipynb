```python
#  ,  ,  ()()

import csv



def find(maxormin,templist,date):                # >>   
    
    if maxormin:            # 
        
        find_temp=max(templist)               #    
        
    else:                  # 
        
        find_temp=min(templist)
        
    find_idx=templist.index(find_temp)         # /    
    find_date=date[find_idx]
    
    return find_temp,find_date                 #/  ,    
    
    
    
f=open("seoul.csv")  

data=csv.reader(f)        

header=next(data)             #  

maxtemp=[]                         #      
mintemp=[]                         #  
cross=[]                           # 
date=[]                         #   

for row in data:               #    
    
   # if row[-1]!='':               #   

    try:                          #  
        
        mintemp.append(float(row[-2]))  
        maxtemp.append(float(row[-1]))   #     ( )
        
        cross.append(float(row[-1])-float(row[-2]))
        
        date.append(row[0])           #   
    
    except:                      # ( )
        
        pass                  

    
    
maxtemp,maxdate =find(True,maxtemp,date)                   #     

mintemp,mindate=find(False,mintemp,date)

crosstemp,crossdate=find(True,cross,date)


print("    "+str(maxdate)+" "+str(maxtemp)+" .")
print("    "+str(mindate)+" "+str(mintemp)+" .")
print("    "+str(crossdate)+" "+str(crosstemp)+" .")

    

f.close()


```
