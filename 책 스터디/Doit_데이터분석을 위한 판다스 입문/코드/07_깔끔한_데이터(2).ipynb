```python
import pandas as pd
```

## 여러 열을 하나로 정리하기

```python
weather = pd.read_csv("../데이터/weather.csv")
weather.head(4)
```

```python
day = list(weather.columns[:4])
weather_melt = pd.melt(weather,id_vars=day,value_name='temp',var_name='day')
weather_melt.head(5)
```

```python
# pivot_table 

weather_tidy = weather_melt.pivot_table(index= ['id','year','month','day'] , columns='element',values='temp',dropna='False')
weather_tidy.head(6)
```

```python
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat.head()
```

## 중복 데이터 처리하기

열 데이터가 반복되면 따로 관리하는것이 좋다.

```python
billborad =pd.read_csv("../데이터/billboard.csv")
billborad.head(5)
```

```python
idx = list(billborad.columns[:5])
billborad_long = pd.melt(billborad,id_vars=idx, value_name= 'rating',var_name='week')
billborad_long.head(5)
```

```python
billborad_long[billborad_long.track == 'Loser'].head()  # 곡명이 중복데이터가 많다 가수값은 따로 관리하자
```

```python
s = list(billborad_long.columns[:4])
billborad_songs = billborad_long[s]
print(billborad_songs.shape)
```

```python
billborad_songs = billborad_songs.drop_duplicates() # 중복 제거 !!
print(billborad_songs.shape)
```

```python
billborad_songs['id'] = range(len(billborad_songs)) # 노래정보와 주간 순위 데이터 를 합침 
billborad_rating =billborad_long.merge(billborad_songs,on=s)
billborad_rating.head(6)
```
