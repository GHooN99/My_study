```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd 
```

```python
tips = sns.load_dataset("tips")
print(tips.head(2))
```

## seaborn 라이브러리 

matplotlib 기반으로 만든 그래프 그리는 라이브러리

```python
ax = plt.subplots() # 기본틀 
ax = sns.distplot(tips['total_bill']) # distplot 으로 히스토그램 만들기 
ax.set_title('Total Bill hist')  # 그래프 없애려면 displot 에 kde =False 인자를 넣어줌
```

```python
ax = plt.subplots() # 기본틀 
ax = sns.distplot(tips['total_bill'],hist = False,rug = True) # hist 인자로 히스토그램 없애기
ax.set_title('Total Bill hist')   # rug 는 데이터 밀집도
```

```python
#count 그래프 
ax = plt.subplots()
ax = sns.countplot('day',data = tips)
```

```python
# 산점도 그래프 with 회귀선 
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill',y = 'tip', data = tips)
```

```python
joint = sns.jointplot(x = 'total_bill',y = 'tip', data = tips) # 산점도 + 히스토그램 
joint.fig.suptitle('joint plot ')
```

```python
joint = sns.jointplot(x = 'total_bill',y = 'tip', data = tips ,kind ="hex",color='g') # 산점도 + 히스토그램 
joint.fig.suptitle('joint plot ') # 육각 산점도 그래프 kind 색깔 color 
```

```python
# 이차원 밀집도 
ax = plt.subplots()
ax = sns.kdeplot(data = tips["total_bill"],data2=tips['tip'], shade = True ) # shade 는 음영 
ax.set_title('Kernel Density plot ')
```

```python
#바 그래프
# 지정 변수의 평균을 나타냄

ax = plt.subplots()
ax = sns.barplot(x='time',y = 'total_bill',data =tips)
ax.set_ylabel("Average total bill")
```

```python
#박스 그래프 
ax = plt.subplots()
ax = sns.boxplot(x='time',y = 'total_bill',data =tips)
ax.set_ylabel("total bill")
```

```python
# 바이올린 그래프 
ax = plt.subplots() 
ax = sns.violinplot(x='time',y = 'total_bill',data =tips)
ax.set_ylabel("Average total bill")
```

```python
# 관계 그래프
fig = sns.pairplot(tips)
```

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot) # 산점도 
pair_grid = pair_grid.map_lower(sns.kdeplot) # 이차원 밀집도 
pair_grid = pair_grid.map_diag(sns.distplot,rug =True ) # 히스토그램
plt.show()
```
