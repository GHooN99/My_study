```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

### seaborn 으로 다변량 그래프 그리기

```python
tips = sns.load_dataset('tips')
print(tips.head(2))
```

```python
ax = plt.subplots()
ax = sns.violinplot(x='time',y='total_bill',hue ='sex',data = tips ,split =True) #hue 인자에 색 ,  split 은 겹치기
```

```python
# 산점도 관계 그래프 
scatter = sns.lmplot(x='total_bill',y = 'tip',data = tips , hue ='sex',fit_reg =True) #fit_reg 는 회귀선 ?
fig =sns.pairplot(tips,hue = 'sex')
```

### 데이터 프레임으로 그래프 그리기

```python
ax = plt.subplots()
ax = tips['total_bill'].plot.hist()  ## 히스토그램 
```

```python
ax = tips.plot.kde(x='total_bill',y='tip')
```

```python
ax = tips.plot.hexbin(x='total_bill',y='tip', gridsize= 10)
```
