### anscombe's quartet

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib notebook
```

```python
anscombe = sns.load_dataset("anscombe")
print(anscombe.head(3)) #DataFrame 이다
```

```python
dataset1 = anscombe[anscombe['dataset'] == 'I'] # boolean 반환 이용 조건식
print(dataset1.head(2))
```

```python
plt.plot(dataset1['x'],dataset1['y']) # 기본은 선으로 그림
plt.show()
```

```python
plt.plot(dataset1['x'],dataset1['y'],'o') # o는 점으로 표시 
```

```python
dataset2 = anscombe[anscombe['dataset'] == 'II']
dataset3 = anscombe[anscombe['dataset'] == 'III']
dataset4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
fig = plt.figure()
axe1 = fig.add_subplot(2,2,1) # 2행2열 1번쨰
axe2 = fig.add_subplot(2,2,2)
axe3 = fig.add_subplot(2,2,3)
axe4 = fig.add_subplot(2,2,4)
axe1.plot(dataset1['x'],dataset1['y'],'o',color = 'r')
axe2.plot(dataset2['x'],dataset2['y'],'o',color = 'g')
axe3.plot(dataset3['x'],dataset3['y'],'o',color = 'b')
axe4.plot(dataset4['x'],dataset4['y'],'o',color = 'y') # 축 ,색 지정 
axe1.set_title('dataset1') # 제목 
axe2.set_title('dataset2')
axe3.set_title('dataset3')
axe4.set_title('dataset4')
fig.suptitle("Anscombe Data") # 전체 제목
fig.tight_layout() # 레이아웃 안겹치게
```

### Tips data

```python
tips = sns.load_dataset("tips") # 식당 팁 지불 데이타
print(tips.head(5)) #나중에 분류 문제로도 해보자 ! 
print(len(tips))
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.set_title("Hist of bill") 
axes1.set_ylabel("total_bill")
axes1.set_xlabel("Frequency")
axes1.hist(tips['total_bill'],bins =10) # hist 로 bin은 x 축 간격
```

```python
# 산점도 이변수 그래프
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.set_title("scatter of bill") 
axes1.set_xlabel("total_bill")
axes1.set_ylabel("Tips")
axes1.scatter(tips['total_bill'],tips['tip']) 
```

```python
# 이산 변수 박스그래프(boxplot)
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.set_title("Boxplot ") 
axes1.set_xlabel("gender")
axes1.set_ylabel("Tip")
axes1.boxplot([tips[tips['sex']=='Male']['tip'],
              tips[tips['sex']=='Female']['tip']],
              labels = ['Male','Female']
             )  # 해당하는 리스트반환
```

```python
# 다변량 산점도 그래프 
def recode_sex(sex):
    if sex =='Male':
        return 1
    else:
        return 0

tips['sex_color'] = tips['sex'].apply(recode_sex) #함수 브로드 케스팅 해서 추가 
print(tips.head(2))
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
axes1.set_title("scatter of bill") 
axes1.set_xlabel("total_bill")
axes1.set_ylabel("Tips")
axes1.scatter(tips['total_bill'],tips['tip'],s=tips['size']*25,c = tips['sex_color'],alpha =0.55)
 # s 는 size c 는 color alpha 는 투명도 
```
